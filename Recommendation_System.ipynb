```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#loading the data
movies_data = pd.read_csv('movies.csv')
movies_data.head()
```

```python
#selecting the relevent features
selected_features = ['genres','keywords','tagline','overview','director']
print(selected_features)
```

# New Section

```python
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
combined_features = movies_data['genres']+''+movies_data['keywords']+''+movies_data['tagline']+''+movies_data['overview']+''+movies_data['director']
```

```python
print(combined_features)
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie_name = input("Enter Your Favorite Movie Name: ")
```

```python
list_of_all_titles = movies_data['title'].tolist()
```

```python
print(list_of_all_titles)
```

```python
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
close_match=find_close_match[0]
print(close_match)
```

```python
index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
```

```python
print(similarity_score)
```

```python
sorted_similar = sorted(similarity_score, key= lambda x:x[1], reverse = True)
print(sorted_similar)
```

```python
print('Movies Suggested for you : \n')
i=1
for movie in sorted_similar:
  index = movie[0]
  title_Suggestion = movies_data[movies_data.index==index]['title'].values[0]
  if (i<20):
    print(i,')',title_Suggestion)
    i=i+1
```
